```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
```

```python
read_file=pd.read_csv('car.csv')
read_file.head()
```

## Select features

```python

X = read_file['enginesize'].values
y = read_file['price'].values
```

```python
X = X.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2 ,random_state=100)
```

```python
x_train.shape # 80% of X data
```

```python
x_test.shape # 20% of X data
```

```python
plt.scatter(x_train, y_train, color='blue')
plt.xlabel('engine size')
plt.ylabel('price')
plt.title('Training Data')
plt.show()
```

## create the model and train it

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

## testing

```python
y_predict = model.predict(x_test)
```

```python
y_test
```

```python
y_predict
```

```python
new_engine_size = np.array([120]).reshape(-1,1)
new_engine_size
```

```python
model.predict(new_engine_size)
```

```python
new_engine_size = np.array([140]).reshape(-1,1)
```

```python
model.predict(new_engine_size)
```

## Accuracy of model

```python
model.score(x_train, y_train)*100
```

```python
model.score(x_test, y_test)*100
```

### Plotting the predicted output vs. the test output

```python
plt.scatter(x_train, y_train, color='blue')
plt.plot(x_test, y_predict, color='red')
plt.xlabel('engine size')
plt.ylabel('price')
plt.title('Training Data vs Model')
plt.show()
```
